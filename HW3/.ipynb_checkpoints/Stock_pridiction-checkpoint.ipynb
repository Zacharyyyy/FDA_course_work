## Data
* https://www.sharecast.com/index/SP_500/prices/download

    * test: 02-Jan-2018 to 31-Dec-2018

```python
import pandas as pd
import numpy as np
import random as rnd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
train_df.head()
```

```python
#plot
plt.figure(figsize=(16,8))
plt.plot(train_df['Close Price'], label='Close Price history')
```

```python
# mean
train_df['mean'] = train_df['Close Price']
test_df['mean'] = test_df['Close Price']

for i in range(2264):
    train_df['mean'][i] = np.mean(train_df['Close Price'][:i])
for i in range(252):
    test_df['mean'][i] = np.mean(test_df['Close Price'][:i])
```

```python
train_df['mean10'] = train_df['Close Price'].rolling(10).mean()
test_df['mean10'] = test_df['Close Price'].rolling(10).mean()

plt.figure(figsize=(15,7))
plt.plot_date(train_df['Date'], train_df['Close Price'], '-')
plt.plot_date(train_df['Date'], train_df['mean10'], '-')
plt.legend(['Close Price', '10days mean'])
plt.xlabel('year')
plt.show()
```

```python
train_df['Date'] = pd.to_datetime(train_df['Date'])
train_df['year'] = train_df['Date'].dt.year
train_df['month'] = train_df['Date'].dt.month
train_df['day'] = train_df['Date'].dt.day
train_df['day_of_week'] = train_df['Date'].dt.dayofweek

test_df['Date'] = pd.to_datetime(test_df['Date'])
test_df['year'] = test_df['Date'].dt.year
test_df['month'] = test_df['Date'].dt.month
test_df['day'] = test_df['Date'].dt.day
test_df['day_of_week'] = test_df['Date'].dt.dayofweek
```

```python
target = np.zeros(len(test_df))
for index in range(len(test_df)-1):
    cur = test_df.iloc[index, 2]
    nxt = test_df.iloc[index+1, 2]
    
    if nxt >= cur:
        target[index] = 1
    else:
        target[index] = 0
        
test_df['target'] = target

target = np.zeros(len(train_df))
for index in range(len(train_df)-1):
    cur = train_df.iloc[index, 2]
    nxt = train_df.iloc[index+1, 2]
    
    if nxt >= cur:
        target[index] = 1
    else:
        target[index] = 0
        
train_df['target'] = target
train_df.head()
```

```python
train_df =train_df['mean10'].fillna(900)
test_df = test_df['mean10'].fillna(900)

train_x = train_df[['Open Price', 'Close Price', 'High Price', 'Low Price', 'Volume', 'mean10']]
train_y = train_df['target']
test_x = test_df[['Open Price', 'Close Price', 'High Price', 'Low Price', 'Volume', 'mean10']]
test_y = test_df['target']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logi_model = LogisticRegression()
logi_model.fit(train_x, train_y)
    
logi_pred = logi_model.predict(test_x)
print(accuracy_score(test_y, logi_pred))
```

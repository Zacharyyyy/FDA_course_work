```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Reviews.csv')  
```

```python
data = df[:10000]
```

```python
data
```

```python
df = data.groupby(['UserId', 'ProfileName'])['Score'] \
            .count() \
            .reset_index(name='Score count') \
            .sort_values(['Score count'], ascending=False)
mean_df = data.groupby(['UserId', 'ProfileName'])['Score'] \
            .mean() \
            .reset_index(name='Score mean') \
            .sort_values(['Score mean'], ascending=False)
df['Score mean'] = mean_df['Score mean']
df.head(10)
```
